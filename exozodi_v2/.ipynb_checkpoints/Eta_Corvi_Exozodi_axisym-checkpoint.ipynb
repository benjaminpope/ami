# To create model images there are a series of steps:
    - Define dust surface density distribution through a function.
    - Define the model parameters.
    - Create input files for RAMC3D: starmodel, spherical grid where model is defined, wavelength grid for thermal monte carlo, dust opacities and dust density. 
    - Run thermal monte carlo to obtain dust temperatures.
    - Create SED or images, etc, based on the pre-computed temperature.

```python
import numpy as np
import disc2radmc 
import matplotlib.pyplot as plt
import astropy.io.fits as pyfits
import json
from scipy import interpolate
```

We first define the parametric function that will set the surface density distribution as a function of r and phi.
These functions need to be able to take numpy arrays as inputs for r and phi.

```python

# For example.
# def Sigma_dust(r, phi, rc, width):
#     sigr=width/2.335 # fwhm to sigma
#     return np.exp( -0.5 * ((r-rc)/sigr)**2.)

# or
def Sigma_dust(r, phi, rmin, rmax, gamma):
    sigma=(r/rmin)**gamma
    sigma[r<rmin]=0.0
    sigma[r>rmax]=0.0
    return sigma
```

We know define the model parameters. These are many, but most of them are fairly obvious. Note that the star is defined using a BT model which needs an input file, and thus you probably want to define that as it is to avoid having to calculate it again.

```python
########################
###### PARAMETERS ######
########################

## STAR
Rstar=1.58 # [Solar radii]
Tstar= 6880 # [K] If this is negative, the code will consider the star as a blackbody. For a realistic stellar model, you can download bt-settl models and indicate their directory later
dir_stellar_templates='/Users/Sebamarino/Astronomy/Stellar_templates/BT-Settl/bt-settl/'
dpc=18.3  # [pc] distance to source 
target='eta_corvi_exozodi_axisym'# name
RA=188.0156449228337    # deg
DEC=-16.196264029437497 # deg

## DUST
Mdust=2.5e-6      # [Mearth] Total dust mass needs to match 22.7+-5.4 % (Lebreton+2016)
# rc=0.7        # [au]
# width=rc*0.5       # [au]
# par_sigma=(rc, width) # list containing the parameters that define the dust surface density. They must be in the same order as in the definition of Sigma_dust
r0=0.8 # au
r1=10. # au
gamma=-1.0
par_sigma=(r0, r1, gamma) # list containing the parameters that define the dust surface density. They must be in the same order as in the definition of Sigma_dust
# values based on Lebreton+2016
amin=0.6       # [mu]  minimum grain size (1 by default)
amax=1.0e3     # [mu]  maximum grain size (1e4 by default)
N_species=10    #  (1 by default) Number of dust size bins to use for the radiative transfer calculations. 
slope = -4.0   #  (-3.5 by default) Slope of the size distribution. This is used for computing opacities and for the mass distribution of each size bin
h=0.05         # vertical aspect ratio =H/r, where H is the vertical standard deviation. This is a constant, but other parametrizations are possible and will be shown below. 

## SPHERICAL GRID WHERE OUR MODEL IS DEFINED
rmin=r0*0.9 # [au] make sure it is small enough to sample the surface density
rmax=r1*1.1 # [au] 
Nr=30    # radial cells (linearly or logspaced)
Nphi=100 # azimuthal cells
Nth=10   # polar angle cells
thmax=np.arctan(h)*5 # maximum polar angle to sample as measured from the midplane.
axisym=False # Consider the disc to be axisymmetric to speed up calculations? it can overwrite Nphi if True and set it to 1
mirror=False  # Mirror the upper half to speed up calculations. This is incompatible with anisotropic scattering.
logr=True # Sample r logarithmically or linearly

# WAVELENGTH GRID (grid to sample the stellar flux in temperature calculations, see radmc3d manual)
lammin=0.09  # [mu]  minimum wavelength to consider in our model (important for temperature calculation)
lammax=1.0e4 # [mu] minimum wavelength to consider in our model (important for temperature calculation)
Nlam=300     # number of cells logarithmically spaced to sample the wavelength range.

# IMAGE PARAMETERS
Npix=512  # number of pixels
dpix=0.001 # pixel size in arcsec
inc=35.    # inclination
PA=117.    # position angle
scattering_mode=2 # scattering mode (0=no scattering, 1=isotropic, 2=anisotropic using H&G function)
```

Now we create the input files that RADMC3D needs based on these parameters. Note that I've commented the lines that define the star since I did this already and doesn't need to be repeated.

```python
### PHYSICAL GRID
gridmodel=disc2radmc.physical_grid(rmin=rmin, rmax=rmax, Nr=Nr, Nphi=Nphi, Nth=Nth, thmax=thmax, mirror=mirror, logr=logr, axisym=axisym)
gridmodel.save()

### WAVELENGTH GRID
lammodel=disc2radmc.wavelength_grid(lammin=lammin, lammax=lammax, Nlam=Nlam)
lammodel.save()

### STAR
# starmodel=disc2radmc.star(lammodel, Tstar=Tstar, Rstar=Rstar,
#                             dir_stellar_templates=dir_stellar_templates # necessary if Tstar>0
#                                )
# starmodel.save()
```

We now define the dust opacities. In this case, grains are 100% astrosilicates. Note that I've commented the line that computes the opacities since I did this already and doesn't need to be repeated. It also requires of another file to use Mie Theory to compute the opacities.

```python
### DUST SIZE DISTRIBUTION AND OPACITY
# path to optical constants that can be found in https://github.com/SebaMarino/disc2radmc/tree/main/opacities/dust_optical_constants
path_opct='/Users/Sebamarino/Astronomy/Codes/disc2radmc/opacities/dust_optical_constants/' 
lnk_files=[path_opct+'astrosilicate_ext.lnk',
           path_opct+'ac_opct.lnk',
           path_opct+'ice_opct.lnk']
densities=[3.5, 3., 1.] # densities in g/cm3
mass_weights=[100.0, 0., 0.] # mixing ratios by mass
dust=disc2radmc.dust(lammodel,
                        Mdust=Mdust,
                        lnk_file=lnk_files,
                        densities=densities,
                        N_species=N_species,
                        slope=slope,
                        N_per_bin=100, # number of species per size bin to have a good representation 
                        mass_weights=mass_weights,
                        tag='mix', # name to give to this new mixed species
                        compute_opct=False) 
#dust.compute_opacities()
```

Now we save the dust density distribution

```python
### DUST DENSITY DISTRIBUTION
dust.dust_densities(grid=gridmodel,function_sigma=Sigma_dust, par_sigma=par_sigma, h=h)
dust.write_density()
```

And finnally, we run RADMC3D to calculate the temperature. Note that there are several parameters that need to be defined. nphot sets how many photon packages to use to calculate the temperature. It should be high enough such that the temperature field doesn't have noise (this is checked later). nphot_scat sets the number of photon packages used to estimate the scattered light flux when calculating images. nphot_mono sets the number of photon packages used to estimate the scattered light flux when calculating SED's or spectra.

```python
sim=disc2radmc.simulation(nphot=100000000, # number of photon packages for thermal monte carlo
                        nphot_scat=100000, # number of photon packages for image
                        nphot_spec=10000,   # number of photon packages for spectrum
                        nphot_mono=10000,   # number of photon packages for the monochromatic monte carlo
                        scattering_mode=scattering_mode, 
                        incl_lines=0, # whether to include gas lines (1) or not (0)
                        modified_random_walk=0, # for very optically thick medium, this is a useful approximation
                        istar_sphere=0, # consider the star a sphere or a point.
                        tgas_eq_tdust=1, # gas temperature equal to the temperature of the first dust species
                        setthreads=4,
                        verbose=True, 
                        )

### RUN MCTHERM to compute temperature
sim.mctherm()
```

```python
# check temperature convergence. If it looks noisy, increase nphot above
Ts=np.fromfile('./dust_temperature.bdat', count=gridmodel.Nr*gridmodel.Nphi*gridmodel.Nth+4, dtype=float)[4:].reshape( (gridmodel.Nphi, gridmodel.Nth, gridmodel.Nr))
plt.pcolormesh(gridmodel.redge, gridmodel.thedge, Ts[0,:,:])
plt.xlabel('Radius [Rsun]')
plt.xscale('log')
plt.ylabel(r'Polar angle [rad]')
plt.colorbar(label='Temperature [K]')
plt.show()
```

```python
# make sed
wavelengths=np.logspace(0.5,1.5, 50)  # um
sed=sim.simsed(wavelengths=wavelengths, dpc=dpc, inc=inc)

# load stellar spectrum
star_lowr=np.loadtxt('stars.inp', skiprows=3+lammodel.Nlam)*1.0e23/dpc**2. # flux in cgs at 1pc to flux in jy at dpc
fstar=interpolate.interp1d(lammodel.lams, star_lowr)
star=fstar(sed[:,0])

## plot sed
plt.plot(sed[:,0], sed[:,1], color='C0', lw=5, label='Model')
plt.plot(sed[:,0], star, color='grey', label='Star')
plt.plot(sed[:,0], sed[:,1]-star, color='C1', lw=5, label='Dust')

plt.xscale('log')
plt.yscale('log')
plt.ylim(1.0e-2, 1.0e1)
plt.xlim(3.0, 30.0)

## add photometry by Grant 

d = json.load(open('sed_grant_etacorvi_2.json','r'))
### photometry
wav_p=d['phot_wavelength'][0]
fnu_p=d['phot_fnujy'][0]
ignore_p=d['phot_ignore'][0]
upper_lim=d['phot_upperlim'][0]

## Photometry
for i in range(len(wav_p)):
    if upper_lim[i] and ignore_p[i]==False:
        plt.plot(wav_p[i],fnu_p[i], "v", color='black')
    elif ignore_p[i]==False:
        plt.plot(wav_p[i],fnu_p[i], 'o', color='black', ms=7.0, alpha=0.5)
        

# BB fit models from Grant
# star
wav_s= np.array(d['star_spec']['wavelength'])
fnu_s= np.array(d['star_spec']['fnujy'])

# disk
wav_d= np.array(d['disk_spec']['wavelength'])
fnu_d= np.array(d['disk_spec']['fnujy'])

wav_hot=np.array(d['model_spectra'][2]['wavelength'])
wav_cold=np.array(d['model_spectra'][1]['wavelength'])

fnu_hot=np.array(d['model_spectra'][2]['fnujy'])
fnu_cold=np.array(d['model_spectra'][1]['fnujy'])


# ### models 
# plt.plot(wav_s,fnu_s, color='C1', ls='--', lw=2.0)
# plt.plot(wav_d,fnu_d+fnu_s, color='C7', lw=3.0, alpha=1.0)

plt.plot(wav_hot,fnu_hot, color='red', lw=2.0, label='Best BB fit from Grant')
plt.plot(wav_hot,fnu_cold, color='blue', lw=2.0)
plt.legend(loc=1)
```

```python
### MAKE IMAGE at 11um
sim.simimage(dpc=dpc, 
            imagename='11um', 
            wavelength=11.0,
            Npix=Npix,
            dpix=dpix,
            inc=inc,
            PA=PA,
            tag=target,
            X0=RA,
            Y0=DEC)
```

```python
fit1=pyfits.open('./images/image_11um_'+target+'.fits')
image=fit1[0].data[0,0,:,:]
total_flux=np.sum(image) # Jy
stellar_flux= np.nanmax(image) # Jy
print('Total flux = %1.3f mJy'%(total_flux*1.0e3))
print('Star flux = %1.3f mJy'%((stellar_flux)*1.0e3))
print('Disc flux = %1.3f mJy'%((total_flux-stellar_flux)*1.0e3))

# mask star
ijstar=np.unravel_index(image.argmax(), image.shape)
image[ijstar]=0.0

fig=plt.figure(figsize=(6,6))
ax1=fig.add_subplot(111)
ax1.pcolormesh(image)
ax1.set_aspect('equal')
```
